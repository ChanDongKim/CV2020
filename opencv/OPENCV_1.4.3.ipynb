# Translation

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('sIMG_2164.jpg')
tx = 100
ty = 150
rows,cols, _ = img.shape
```

```python
M = np.float32([[1,0,tx],[0,1,ty]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.rectangle(dst, (0,0), (tx,ty), (0, 255, 0), -1)
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Rotation

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('sIMG_2164.jpg')

rows, cols, _ = img.shape
cx = 200
cy = 200
theta = 30
M = cv2.getRotationMatrix2D((cx,cy),theta,1)
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.circle(dst, (cx,cy),10, (0, 0, 255), -1)
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#  Affine Transformation

```python
img = cv2.imread('dsu4.jpg')
```

```python
rows,cols,ch = img.shape

pts1 = np.float32([[235,457],[238,269],[569,231]])
pts2 = np.float32([[235,457],[235,270],[570,270]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

#  Perspective Transformation

```python
import cv2
import numpy as np

img = cv2.imread('dsu5.jpg')

rows,cols,ch = img.shape



pts1 = np.float32([[583,489],[578,336],[753,377],[759,502]])
pts2 = np.float32([[583,489],[583,336],[753,336],[753,489]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
